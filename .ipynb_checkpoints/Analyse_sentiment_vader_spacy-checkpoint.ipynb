```python
import pandas as pd
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('analyse_sentiment_facebook.csv')
```

```python
df.head()
```

```python
model = SentimentIntensityAnalyzer()
```

```python
def get_sentiment(text):
    scores = model.polarity_scores(text)
    return scores.get('compound')
```

```python
%time df['A'] = np.vectorize(get_sentiment)(df.Avis)
```

```python
df[''] = df['retweets'].replace(np.nan, df['retweets'].mean)
```

```python
df.to_csv('trump_sentiment.csv')
```

```python
df['is_positive'] = df.sentiment>0
```

```python
import spacy
import re
```

```python
nlp = spacy.load('en_core_web_sm')

def get_entities(text):
    doc = nlp(text)
    list_ = []
    for ent in doc.ents:
        if ent.label_.lower() not in ('time', 'money', 'date'):
            temp_  = re.sub('[^a-z0-9]+', '_', ent.text.lower())
            list_.append('__'.join([temp_, ent.label_.upper()]))
    return ''.join(list_)
```

```python
%time df['entities'] = np.vectorize(get_entities)(df.content)
```
